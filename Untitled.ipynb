```python
#今回の学習の狙いはAPIkeyを用いてシステム内にAIを呼び出すことです。コーディングは全てAIに任せ、質問し、グラフの箇所以外は理解しました。特に、GPTの出力をプログラムが直接読み込めるjson形式に指定したり出来ることは自分の中でかなり衝撃的でもうこういったシステム内に巨大なデータベースを用意することはいらないんじゃないかなと思ったりしました。
import json
import matplotlib.pyplot as plt
from openai import OpenAI
import getpass  # パスワード入力用のライブラリを追加

# 1. input形式でAPIキーを埋め込む。最初AIに提案されたコードは環境変数を用いるものだったが、前回の壊滅的な血糖値システムからinput形式を学んだのでセキュリティ上も問題ないためそちらに変更した。
api_key = getpass.getpass("OpenAIのAPIキーを入力してください (画面には表示されません): ")
client = OpenAI(api_key=api_key)

# 2. ユーザーからの入力を受け取る
food_name = input("調べたい食材を入力してください: ")
print(f"「{food_name}」のデータをAIで検索中...")

# 3. OpenAI APIに送る「指示書（プロンプト）」を組み立てる
prompt = f"""
ユーザーが入力した食材「{food_name}」について、信頼できる医療・栄養学のデータを基に、以下の3つの数値を調査してください。

1. gi: その食材のGI値（一般的な数値）
2. peak: 血糖値がピーク（頂点）に達するまでの時間（分）
3. down: 血糖値が元の正常値に戻るまでの時間（分）

必ず以下のJSONフォーマットの形式「のみ」で出力してください。挨拶や余計な解説のテキストは絶対に含めないでください。
{{
    "gi": 数値,
    "peak": 数値,
    "down": 数値
}}
"""

# 4. AIにリクエストを送信し、結果を受け取る
response = client.chat.completions.create(
    model="gpt-4o-mini",  # コスパが良く高速なモデルを指定
    messages=[{"role": "user", "content": prompt}],
    response_format={"type": "json_object"}  # 確実にJSON形式で返させる強制機能
)

# 5. AIの返答（文字列）を、Pythonで扱える「辞書型」に変換する
ai_output_string = response.choices[0].message.content
data = json.loads(ai_output_string)

# 6. 辞書型から各数値を引っこ抜く
gi = data["gi"]
peak = data["peak"]
down = data["down"]

print(f"\n--- 検索結果 ---")
print(f"GI値: {gi} / ピーク時間: {peak}分 / 下がりきる時間: {down}分")

# 7. ライブラリ（Matplotlib）を使ってグラフのデータを組み立てる
# X軸（時間）：[食前, ピーク時, 正常復帰時]
time_axis = [0, peak, down]
# Y軸（血糖値の上昇幅）：[0から始まり, GI値まで上がり, 0に戻る]
glucose_axis = [0, gi, 0]

# 8. グラフをお絵描きする設定
plt.figure(figsize=(8, 5))  # グラフのウインドウサイズを設定（横8インチ、縦5インチ）
plt.plot(time_axis, glucose_axis, marker='o', color='red', linewidth=2, label="血糖値変動")

# グラフの見栄えを整える設定
plt.title(f"Predictive Glucose Spike: {food_name}", fontsize=14)
plt.xlabel("Time (minutes)", fontsize=12)
plt.ylabel("Glucose Increase Level", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)  # グリッド線（点線）をうっすら入れる
plt.axhline(70, color='orange', linestyle=':', label="High GI Threshold (70)") # 危険ラインの横線
plt.legend()  # 凡例（ラベルの説明）を表示

# 9. 気分の傾向をロジックで判定してターミナルに出力
print("\n--- 体調・気分の傾向予測 ---")
if gi >= 70:
    print("⚠️ 警告: シュガースパイク（血糖値の急上昇）が発生します！")
    print(f"食後 {peak} 分頃までは一時的に脳がハイになりますが、{down} 分に向けて急降下します。")
    print("その際、激しい眠気、集中力の低下、イライラ感に襲われる可能性が非常に高いです。")
else:
    print("✅ 安定: 血糖値は緩やかに推移します。")
    print(f"食後 {down} 分後まで安定してエネルギーが脳に供給され続けます。")
    print("メンタルが安定し、仕事や勉強のパフォーマンスを高く維持できる理想的な食事です。")

# 10. 最後にグラフを画面に表示する

plt.show()
```
